```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters
num_steps = 10  # Number of steps to animate
step_size = 0.1  # Step size for each dimension

# Function to plot vectors
def plot_vectors(ax, vectors):
    ax.cla()  # Clear current plot

    for vector in vectors:
        ax.quiver(0, 0, 0, *vector, color='b', arrow_length_ratio=0.1)

    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Sliding Vectors in 3D Space')

    plt.draw()
    plt.pause(0.1)

# Create initial vectors
vectors = [
    np.array([1, 0, 0]),
    np.array([0, 1, 0]),
    np.array([0, 0, 1])
]

# Initialize plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Animate sliding vectors
for i in range(num_steps):
    vectors[0][0] += step_size
    vectors[1][1] += step_size
    vectors[2][2] += step_size

    plot_vectors(ax, vectors)

plt.show()
```
