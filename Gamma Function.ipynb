Γ(x) = ∫[0, ∞] t^(x-1) * e^(-t) dt 

```python
from scipy.special import gamma

def integrate_gamma(x):
    return gamma(x)

import numpy as np

x = np.linspace(0.5, 2, 100)
y = [integrate_gamma(i) for i in x]

import matplotlib.pyplot as plt

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Gamma(x)')
plt.title('Gamma Function')
plt.show()
```
